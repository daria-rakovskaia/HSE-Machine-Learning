```python
import pandas as pd
import numpy as np
from apyori import apriori
```

```python
df = pd.read_csv("foods_.csv")
df.columns = ["item_" + str(i + 1) for i in range(len(df.columns))]
df = df.replace("ice_crea", "ice_cream")
df.head()
```

```python
transactions = [list(value) for value in df.values]

rules = apriori(transactions, min_support=0.1, min_confidence=0.25, min_lift=1.1,
                min_length=2, max_length=2)

result = pd.DataFrame(columns=["items_base", "items_add",
                               "support", "confidence", "lift"])

for rule in rules:
  result.loc[len(result)] = [list(rule[2][0][0])[0], list(rule[2][0][1])[0],
                             rule.support, rule[2][0][2], rule[2][0][3]]

result = result.sort_values(by="lift", ascending=False)
result[:10]
```
