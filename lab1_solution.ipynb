```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objs as go
```

```python
df = pd.read_csv("winequality-red.csv", delimiter=';')
```

```python
min_quality_df = df.loc[(df["quality"] == min(df["quality"]))]
max_quality_df = df.loc[(df["quality"] == max(df["quality"]))]
```

```python
plt.plot(min_quality_df["alcohol"], 'o-b', label="worst quality wines")
plt.plot(max_quality_df["alcohol"], 'o-g', label="best quality wines")
plt.title("Indicators of the percentage of alcohol in red wines of the best and worst quality")
plt.xlabel("Wines")
plt.ylabel("Percentage of alcohol")
plt.legend()
plt.grid(True)
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(y=min_quality_df["alcohol"], name="worst quality wines"))
fig.add_trace(go.Scatter(y=max_quality_df["alcohol"], name="best quality wines"))
fig.update_layout(legend_orientation="v",
                  title="Indicators of the percentage of alcohol in red wines of the best and worst quality",
                  xaxis_title="Wines",
                  yaxis_title="Percentage of alcohol")
```

```python
plt.hist(df["pH"], 15)
plt.grid(axis='y')
plt.title("pH level (describes how acidic or basic a wine is)")
plt.xlabel("pH")
plt.ylabel("Amount")
plt.show()
```

```python
fig = go.Figure(data=[go.Histogram(x=df["pH"], nbinsx=15, marker_color='#7A202C')])
fig.update_layout(title="pH level (describes how acidic or basic a wine is)",
                  xaxis_title="pH",
                  yaxis_title="Amount")
```

```python
group_df = df.groupby("quality").groups
labels, values = [], []
for key in group_df:
    labels.append("quality " + str(key))
    values.append(group_df[key].size)
explode = [1, 0, 0, 0, 0, 1]
plt.pie(values, 
        autopct='%.2f%%',
        explode=explode)
plt.title("Overall quality values")
plt.legend(labels, 
           loc="center right",
           bbox_to_anchor=(1.2, 0, 0.5, 1))
plt.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Pie(values=values, labels=labels))
fig.update_layout(title="Overall quality values")
```

```python
plt.boxplot(df["quality"])
plt.title("Quality boxplot")
plt.ylabel("Quality")
plt.grid(axis='y')
plt.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Box(y=df["quality"], name="Boxplot for quality"))
fig.update_layout(title="Quality boxplot",
                  yaxis_title="Quality")
```
