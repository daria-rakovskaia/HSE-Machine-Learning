```python
import pandas as pd
```

```python
df = pd.read_csv("poems.csv")
df = df.drop("poem", axis=1)
df = df.dropna()
df.head()
```

```python
df.groupby(["writer"]).count().sort_values(by="text", ascending=False)
```

```python
group_1 = df.groupby(["writer"]).get_group("Брюсов Валерий Яковлевич")["text"].tolist()
group_0 = df.groupby(["writer"]).get_group("Фет Афанасий Афанасьевич")["text"].tolist()
group_0 += df.groupby(["writer"]).get_group("Пушкин Александр Сергеевич")["text"].tolist()

res_1 = [1 for i in range(len(group_1))]
res_0 = [0 for i in range(len(group_0))]
```

```python
from sklearn.utils import shuffle
```

```python
data = pd.DataFrame({"text": group_1 + group_0,
                     "result": res_1 + res_0})

data = shuffle(data, random_state=0).reset_index(drop=True)
data.head()
```

```python
import ru_core_news_lg
nlp = ru_core_news_lg.load()
```

```python
data["pos"] = data["text"].apply(lambda x: nlp(x))
```

```python
data["pos_names"] = data["pos"].apply(lambda x: [token.tag_ for token in x])
```

```python
data.head()
```

```python
pos = set()
for item in data["pos_names"].values:
  pos.update(item)
```

```python
for p in pos:
  data[p] = data["pos_names"].apply(lambda x: x.count(p))
```

```python
data["pos_sum"] = [sum(data.iloc[i][4:]) for i in range(len(data))]
```

```python
for i in range(5):
  for p in pos:
    data[p][i] = data.iloc[i][p] / data.iloc[i]["pos_sum"]
```

```python
data.head()
```

```python
res_df = data.drop(["text", "pos", "pos_names", "pos_sum"], axis=1)
res_df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
scaled_features = scaler.fit_transform(res_df)
scaled_data = pd.DataFrame(data=scaled_features, columns=res_df.columns)
scaled_data.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score
```

```python
x_train, x_test, y_train, y_test = train_test_split(scaled_data.drop(["result"], axis=1), scaled_data["result"],
                                                    test_size=0.2, random_state=42)
```

```python
clf = SVC()
clf.fit(x_train, y_train)
y_prediction = clf.predict(x_test)
```

```python
print("SVM", "\n")
print("precision_score:", precision_score(y_test, y_prediction) * 100, '%')
print("recall_score:", recall_score(y_test, y_prediction) * 100, '%')
print("f1_score:", f1_score(y_test, y_prediction) * 100, '%')
print("roc_auc_score:", roc_auc_score(y_test, y_prediction) * 100, '%')
```

```python
clf = RandomForestClassifier()
clf.fit(x_train, y_train)
y_prediction = clf.predict(x_test)
```

```python
print("RandomForestClassifier", "\n")
print("precision_score:", precision_score(y_test, y_prediction) * 100, '%')
print("recall_score:", recall_score(y_test, y_prediction) * 100, '%')
print("f1_score:", f1_score(y_test, y_prediction) * 100, '%')
print("roc_auc_score:", roc_auc_score(y_test, y_prediction) * 100, '%')
```

```python
clf = GaussianNB()
clf.fit(x_train, y_train)
y_prediction = clf.predict(x_test)
```

```python
print("GaussianNB", "\n")
print("precision_score:", precision_score(y_test, y_prediction) * 100, '%')
print("recall_score:", recall_score(y_test, y_prediction) * 100, '%')
print("f1_score:", f1_score(y_test, y_prediction) * 100, '%')
print("roc_auc_score:", roc_auc_score(y_test, y_prediction) * 100, '%')
```
